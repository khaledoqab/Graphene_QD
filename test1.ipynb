```python
import numpy as np
import matplotlib.pyplot as plt

data = np.array([
    [0.375, 0.91],
    [0.297, 0.84],
    [0.212, 0.77],
    [0.128, 0.676],
    [0.041, 0.56],
    [-0.048, 0.446],
    [-0.135, 0.355],
    [-0.224, 0.287],
    [-0.304, 0.2175],
    [-0.384, 0.150],
    [-0.464, 0.0996]
])
plt.figure(figsize=(8, 6))
plt.plot(data[:, 0], data[:, 1], marker='o', linestyle='-', color='b')
plt.title('Data Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.xlim(-0.5, 0.5)
plt.ylim(0, 1)
plt.xticks(np.arange(-0.5, 0.6, 0.1))
plt.yticks(np.arange(0, 1.1, 0.1))
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.title('Plot of Given Data Points')
plt.savefig('data_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Perform a least squares linear fit to the data
coeffs = np.polyfit(data[:, 0], data[:, 1], deg=1)
fit_fn = np.poly1d(coeffs)

# Plot the original data and the fit
plt.figure(figsize=(8, 6))
plt.plot(data[:, 0], data[:, 1], 'bo', label='Data')
plt.plot(data[:, 0], fit_fn(data[:, 0]), 'r-', label='Least Squares Fit')
plt.title('Least Squares Linear Fit')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.legend()
plt.show()
```
